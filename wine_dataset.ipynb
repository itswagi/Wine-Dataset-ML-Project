```python
import pandas as pd
from sklearn.datasets import load_wine
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
wine_data = load_wine()

print(wine_data)
```

```python
print(wine_data.DESCR)
```

```python
print(wine_data.keys())
print(wine_data.feature_names)
```

```python
print(wine_data.target_names)
```

```python
print(wine_data.target)
```

```python
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names, index=None)
print(wine_df.head())
```

```python
wine_df['target'] = pd.DataFrame(wine_data.target)
print(wine_df.head())
```

```python
print(wine_df.isnull().sum())
```

```python
print(wine_df.shape)
```

```python
corr_matrix = wine_df.corr().round(2)
print(corr_matrix)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=corr_matrix, annot=True)
plt.show()
```
