```python
import pandas as pd
from sklearn.datasets import load_wine
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
wine_data = load_wine()

print(wine_data)
```

```python
print(wine_data.DESCR)
```

```python
print(wine_data.keys())
print(wine_data.feature_names)
```

```python
print(wine_data.target_names)
```

```python
print(wine_data.target)
```

```python
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names, index=None)
print(wine_df.head())
```

```python
wine_df['target'] = pd.DataFrame(wine_data.target)
print(wine_df.head())
```

```python
print(wine_df.isnull().sum())
```

```python
print(wine_df.shape)
```

```python
corr_matrix = wine_df.corr().round(2)
print(corr_matrix)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=corr_matrix, annot=True)
plt.show()
```

```python
data_features = wine_df[['flavanoids', 'alcalinity_of_ash']]
print(data_features.head())
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data_features.flavanoids, data_features.alcalinity_of_ash, hue=wine_df.target)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_features, wine_df.target, test_size = 0.25, random_state=1)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression().fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
data_features = wine_df[['flavanoids', 'alcalinity_of_ash', 'proline', 'nonflavanoid_phenols']]
print(data_features.head())
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection = '3d')

x = data_features.flavanoids
y = data_features.alcalinity_of_ash
z = data_features.proline

ax.set_xlabel('Flavanoids')
ax.set_ylabel('Alcalinity_of_ash')
ax.set_zlabel('Proline')

ax.scatter(x, y, z, c=wine_df.target)

plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_features, wine_df.target, test_size = 0.25, random_state=None)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = LogisticRegression().fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
from sklearn.svm import SVC

model = SVC().fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
from sklearn.model_selection import GridSearchCV

param = {
        'C': [0.1, 0.8, 0.9, 1, 1.1, 1.2, 1.3, 1.4],
        'kernel': ['linear', 'rbf'],
        'gamma': [0.1, 0.8, 0.9, 1, 1.1, 1.2, 1.3, 1.4]
    }

svc = SVC()
clf = GridSearchCV(svc, param)
clf.fit(X_train, y_train)
print(clf.best_params_)
```

```python
model = SVC(C=0.9, gamma=0.1, kernel='linear').fit(X_train, y_train)
print(model.score(X_test, y_test))
```
